## Operators

Python provides a variety of operators to perform different types of operations. We already saw some example of use of the '+' and the '*' operators above but is a list the most commonly used ones:


### Arithmetic **Operators**

Arithmetic operators are used to perform basic mathematical operations.

- **Addition (`+`)**: Adds two numbers.
    ```python
    x = 5
    y = 3
    result = x + y  # result is 8
    ```

- **Subtraction (`-`)**: Subtracts one number from another.
    ```python
    result = x - y  # result is 2
    ```

- **Multiplication (`*`)**: Multiplies two numbers.
    ```python
    result = x * y  # result is 15
    ```

- **Division (`/`)**: Divides one number by another, returning a float.
    ```python
    result = x / y  # result is 1.666...
    ```

- **Floor Division (`//`)**: Divides one number by another, returning an integer. **TRUCATES, or ROUNDS DOWN**
    ```python
    result = x // y  # result is 1
    ```

- **Modulus (`%`)**: Returns the remainder of a division.
    ```python
    result = x % y  # result is 2
    ```

- **Exponentiation (`**`)**: Raises one number to the power of another.
    ```python
    result = x ** y  # result is 125
    ```

````python
x=5
y=3

print(x*y/2)

print(x*y//2)

print(x ** y) #x at the power y, in our case 5*5*5
````

### Comparison Operators



Comparison operators are used to compare two values.

- **Equal to (`==`)**: Checks if two values are equal.
    ```python
    x == y  # result is False
    ```

- **Not equal to (`!=`)**: Checks if two values are not equal.
    ```python
    x != y  # result is True
    ```

- **Greater than (`>`)**: Checks if the left value is greater than the right.
    ```python
    x > y  # result is True
    ```

- **Less than (`<`)**: Checks if the left value is less than the right.
    ```python
    x < y  # result is False
    ```

- **Greater than or equal to (`>=`)**: Checks if the left value is greater than or equal to the right.
    ```python
    x >= y  # result is True
    ```

- **Less than or equal to (`<=`)**: Checks if the left value is less than or equal to the right.
    ```python
    x <= y  # result is False
    ```

````python
print(x, y)
````

````python
print(x < y)
print(x == 5)
````

### Logical Operators


Logical operators are used to combine conditional statements. Here are the most common ones (still using `x = 5` and `y = 3`):

- **and**: Returns True if both statements are true.
    ```python
    (x > y) and (x < 10)  # result is True
    ```

- **or**: Returns True if one of the statements is true. **Note this is not exclusive OR (XOR)**.
    ```python
    (x > y) or (x > 10)  # result is True
    ```

- **not**: Reverses the result, returns False if the result is true.
    ```python
    not (x > y)  # result is False
    ```

Note: The parentheses are not required here because of the [priority of operations](https://www.programiz.com/python-programming/precedence-associativity), but they do not hurt and can add to the clarity.

````python
bool_one = x > y
print(bool_one)

bool_two = x < 10
print(bool_two)

print(bool_one and bool_two)
````

````python
print(bool_one and (x < 4))  # True and False ==> False
print(bool_one or (x < 4))   # True or False ==> True
````

````python
var_with_bool = bool_one or (x < 4)
print(not(var_with_bool)) # not True  ==> False
````

## Basic Data Structures

Python offers several built-in data structures essential for organizing and managing data. The most commonly used ones are lists and dictionaries. These are similar to variables in that they store a piece of information (e.g., a number) but can now store many things rather than just one thing. You can think not of a box with one thing in it (variable) but of a container full of boxes (a data structure).

### Lists


A list is an ordered, mutable collection of items. Lists are versatile and can hold items of different data types.

Note: Since it is **mutable**, we can modify it after its creation.

- **Creating a list**: Lists are created using square brackets `[]`.
    ```python
    fruits = ["apple", "banana", "cherry"]
    ```

- **Accessing Elements**: Access elements by their index **(starting from 0)**.
    ```python
    print(fruits[0])  # Output: apple
    ```

- **Modifying Elements**: Change the value of a specific element.
    ```python
    fruits[1] = "blueberry"  # Overwrite "banana" with blueberry
    ```

- **Adding Elements**: Use `append()` to add an element to the end of the list.
    ```python
    fruits.append("date")
    ```

- **Removing Elements by Value**: Use `remove()` to remove a specific element with that value.
    ```python
    fruits.remove("apple")
    ```

- **Removing Elements by Index**: Use `pop()` to remove a specific element at a specific index.
    ```python
    fruits.pop(0)  # Removes the first item of the list.
    ```

````python
my_list = []

print(my_list)
````

````python
my_list.append("luci")
print(my_list,'\n')

my_list.append("marry")
print(my_list,'\n')

#you can add a to a list
my_list.append(["bob","tom"])
print(my_list)
````

````python
# But most of the time, your probably want to "extend" the list by another list...
my_list.remove(["bob","tom"])
print(my_list)
my_list.extend(["bob","tom"])
print(my_list)
````

````python
#getting the last element
last_elem = my_list[-1]
print(last_elem)
````

````python
my_list[0] = 1801
print(my_list,'\n')
````

````python
my_list.remove('marry')
print(my_list)
````

````python
my_list.pop(0)
print(my_list)
````

````python
# The + operator can be used to concatenate two lists
my_list + ["some", "more", "stuff"]
````

### Dictionary (a.k.a. the reason why Python is so great)


A dictionary is an unordered, mutable collection of key-value pairs. Each key must be unique.

- **Creating a Dictionary**: Dictionaries are created using curly braces `{}`, separating keys and values by ":" (i.e., key: value).
    ```python
    person = {"name": "John", "age": 30, "city": "New York"}
    ```

- **Accessing Elements**: Access values by their keys.
    ```python
    print(person["name"])  # Output: John
    ```

- **Modifying Elements**: Change the value associated with a specific key.
    ```python
    person["age"] = 31
    ```

- **Adding Elements**: Add a new key-value pair.
    ```python
    person["job"] = "Engineer"
    ```

- **Removing Elements**: Use `del` to remove a key-value pair.
    ```python
    del person["city"]
    ```


````python
people_colors = {
    "Alice": "blue",
    "Bob": "green",
    "Charlie": "red",
    "Kuba": "Not orange",
    "Cocky": "Garnet"
}

print(people_colors)
````

````python
print('Favorite color of Cocky is:', people_colors["Cocky"])
````

### Tuple

A tuple is an ordered, immutable collection of items. Once created, the elements of a tuple cannot be changed. A great way to think of a tuple is like an immutable list (can't be changed after creation).

- **Creating a Tuple**: Tuples are created using parentheses `()`.
    ```python
    dimensions = (1920, 1080)
    ```

- **Accessing Elements**: Access elements by their index.
    ```python
    print(dimensions[0])  # Output: 1920
    ```

- **Immutability**: You cannot modify elements of a tuple.
    ```python
    # This will raise an error
    dimensions[0] = 1280
    ```

````python
# Tuple unpacking
dimensions = (1920, 1080)  # Tuple
x_dim, y_dim = dimensions
print(x_dim,' ',y_dim)
````

````python
# This works with any iterable (i.e., data structures that queried for one element at a time)
dimensions = [1920, 1080]   # list
x_dim, y_dim = dimensions
print(x_dim,' ',y_dim)
````

````python
# ... even with dictionaries. You then get the "keys" of the dictionary.
dimensions = {"x": 1920, "y": 1080}   # dictionary
x_dim, y_dim = dimensions
print(x_dim,' ',y_dim)
````

### Set


A set is an unordered collection of unique items. Sets are useful for membership testing and eliminating duplicate entries. Just like a set in mathematics.

- **Creating a Set**: Sets are created using curly braces `{}` or the `set()` function.
    ```python
    colors = {"red", "green", "blue"}
    ```
    Don't get confused with `{"red": 1, "green": 2, "blue": 3}` which would created a dictionary, not a set!

- **Adding Elements**: Use `add()` to add an element to the set.
    ```python
    colors.add("yellow")
    ```

- **Removing Elements**: Use `remove()` to remove a specific element.
    ```python
    colors.remove("green")
    ```

- **Set Operations**: Perform operations like union, intersection, and difference.
    ```python
    set1 = {1, 2, 3}
    set2 = {3, 4, 5}
    
    union_set = set1.union(set2)  # {1, 2, 3, 4, 5}
    intersection_set = set1.intersection(set2)  # {3}
    difference_set = set1.difference(set2)  # {1, 2}
    ```


````python
set00 = {2, 8, 90, 2}
print(len(set00))  # Returns 3 because 2 is a duplicated value; the set has only three unique values!
````
