### Cloud Computing

*   What is Cloud Computing?
*   Why Cloud Computing?


**Cloud Computing like Baking:**

Imagine normal coding like this: Coming up with and writing down a recipe is like writing the code. The baking part is like running the code on your computer. And the final baked product is like the output/result of the code.

Cloud Computing: You write the code/recipe, then send it off to be baked by someone else. They come back with the final baked good. In our case, we send off our code, Google runs it, and the code output is sent back to you.

**So what's the advantage?**

Let's say we want to try a new recipe that requires a specific type of oven. We don't have to worry about buying and installing a whole new oven just to try it one time.

Or, what if we had family coming to town and needed to make five times as many cakes as normal, but just once? Since our outsourced baker has many ovens, we can just pay the additional cost of using a few more ovens rather than buying a bunch of new ovens and having to install them all.

## Operators

Python provides a variety of operators to perform different types of operations. We already kind of saw the '+' above but here are the most commonly used ones:



### Arithmetic **Operators**

Arithmetic operators are used to perform basic mathematical operations.

- **Addition (`+`)**: Adds two numbers.
    ```python
    x = 5
    y = 3
    result = x + y  # result is 8
    ```

- **Subtraction (`-`)**: Subtracts one number from another.
    ```python
    result = x - y  # result is 2
    ```

- **Multiplication (`*`)**: Multiplies two numbers.
    ```python
    result = x * y  # result is 15
    ```

- **Division (`/`)**: Divides one number by another, returning a float.
    ```python
    result = x / y  # result is 1.666...
    ```

- **Floor Division (`//`)**: Divides one number by another, returning an integer. **TRUCATES, or ROUNDS DOWN**
    ```python
    result = x // y  # result is 1
    ```

- **Modulus (`%`)**: Returns the remainder of a division.
    ```python
    result = x % y  # result is 2
    ```

- **Exponentiation (`**`)**: Raises one number to the power of another.
    ```python
    result = x ** y  # result is 125
    ```

````python
x=5
y=3

print(x*y/2)

print(x*y//2)

print(x ** y) #x at the power y, in our case 5*5*5

````

- **Greater than (`>`)**: Checks if the left value is greater than the right.

- **Less than (`<`)**: Checks if the left value is less than the right.

- **Greater than or equal to (`>=`)**: Checks if the left value is greater than or equal to the right.
    ```python
    x >= y  # result is True
    ```

- **Less than or equal to (`<=`)**: Checks if the left value is less than or equal to the right.

Logical operators are used to combine conditional statements.

- **and**: Returns True if both statements are true.
    ```python
    (x > y) and (x < 10)  # result is True
    ```

- **or**: Returns True if one of the statements is true. **Note this is not XOR**
    ```python
    (x > y) or (x > 10)  # result is True
    ```

- **not**: Reverses the result, returns False if the result is true.
    ```python
    not (x > y)  # result is False
    ```

Note: The parentheses are not required here because of the [priority of operations](https://www.programiz.com/python-programming/precedence-associativity), but they do not hurt and can add to the clarity.

````python
bool_one = x > y
print(bool_one)

bool_two = x < 10
print(bool_two)

print(bool_one and bool_two)
````

````python
print(bool_one and (x < 4))  # False
print(bool_one or (x < 4))   # True
````

````python
var_with_bool = bool_one or (x < 4)
print(not(var_with_bool)) # not True
````

## Basic Data Structures

Python offers several built-in data structures essential for organizing and managing data. The most commonly used ones are lists and dictionaries. These are similar to variables in that they store a piece of information (e.g., a number) but can now store many things rather than just one thing. You can think not a box with one thing in it (variable) but a container full of boxes (a data structure).

### Lists


A list is an ordered, mutable collection of items. Lists are versatile and can hold items of different data types.

Note: Since it is mutable we can change and modify it after creating.

- **Creating a list**: Lists are created using square brackets `[]`.
    ```python
    fruits = ["apple", "banana", "cherry"]
    ```

- **Accessing Elements**: Access elements by their index **(starting from 0)**.
    ```python
    print(fruits[0])  # Output: apple
    ```

- **Modifying Elements**: Change the value of a specific element.
    ```python
    fruits[1] = "blueberry"  # Overwrite "banana" with blueberry
    ```

- **Adding Elements**: Use `append()` to add an element to the end of the list.
    ```python
    fruits.append("date")
    ```

- **Removing Elements by Value**: Use `remove()` to remove a specific element with that value.
    ```python
    fruits.remove("apple")
    ```

- **Removing Elements by Index**: Use `pop()` to remove a specific element at a specific index.
    ```python
    fruits.pop(0)  # Removes the first item of the list.
    ```

````python
my_list = []

print(my_list)
````

````python
my_list.append("luci")
print(my_list,'\n')

my_list.append("marry")
print(my_list,'\n')

#you can add a to a list
my_list.append(["bob","tom"])
print(my_list)
````

### Dictionary (a.k.a. the reason why Python is so great)


A dictionary is an unordered, mutable collection of key-value pairs. Each key must be unique.

- **Creating a Dictionary**: Dictionaries are created using curly braces `{}`, separating keys and values by ":" (i.e., key: value).
    ```python
    person = {"name": "John", "age": 30, "city": "New York"}
    ```

- **Accessing Elements**: Access values by their keys.
    ```python
    print(person["name"])  # Output: John
    ```

- **Modifying Elements**: Change the value associated with a specific key.
    ```python
    person["age"] = 31
    ```

- **Adding Elements**: Add a new key-value pair.
    ```python
    person["job"] = "Engineer"
    ```

- **Removing Elements**: Use `del` to remove a key-value pair.
    ```python
    del person["city"]
    ```


````python
people_colors = {
    "Alice": "blue",
    "Bob": "green",
    "Charlie": "red",
    "Kuba": "Not orange",
    "Cocky": "Garnet"
}

print(people_colors)
````

### Tuple

A tuple is an ordered, immutable collection of items. Once created, the elements of a tuple cannot be changed. A great way to think of an Tuple is like an immutable list (can't be changed after creation).

- **Creating a Tuple**: Tuples are created using parentheses `()`.
    ```python
    dimensions = (1920, 1080)
    ```

- **Accessing Elements**: Access elements by their index.
    ```python
    print(dimensions[0])  # Output: 1920
    ```

- **Immutability**: You cannot modify elements of a tuple.
    ```python
    # This will raise an error
    # dimensions[0] = 1280
    ```

````python
#Tuple unpacking
dimensions = (1920,1080)
x_dim, y_dim = dimensions
print(x_dim,' ',y_dim)
````

### Set


A set is an unordered collection of unique items. Sets are useful for membership testing and eliminating duplicate entries. Just like a set in mathimatics.

- **Creating a Set**: Sets are created using curly braces `{}` or the `set()` function.
    ```python
    colors = {"red", "green", "blue"}
    ```

- **Adding Elements**: Use `add()` to add an element to the set.
    ```python
    colors.add("yellow")
    ```

- **Removing Elements**: Use `remove()` to remove a specific element.
    ```python
    colors.remove("green")
    ```

- **Set Operations**: Perform operations like union, intersection, and difference.
    ```python
    set1 = {1, 2, 3}
    set2 = {3, 4, 5}
    
    union_set = set1.union(set2)  # {1, 2, 3, 4, 5}
    intersection_set = set1.intersection(set2)  # {3}
    difference_set = set1.difference(set2)  # {1, 2}
    ```
